```python
# Import a library
import pandas as pd
```

```python
# Load dataset
dataset = pd.read_csv('bt1.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Câu 1

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

LINKAGES = ['ward', 'complete', 'average', 'single']
for i in range(len(LINKAGES)):
    dendrogram = sch.dendrogram(sch.linkage(X, method=LINKAGES[i]))
    plt.title(f'Dendogram by {LINKAGES[i]} method')
    plt.xlabel('customers')
    plt.ylabel('Euclidean distance')
    plt.show()
```

## Câu 2

```python
from sklearn.cluster import AgglomerativeClustering

COLORS = ['red', 'green', 'blue', 'yellow', 'purple', 'brown', 'orange', 'lime', 'aqua', 'blue']
```

### ward method

```python
clustering = AgglomerativeClustering(linkage='ward', n_clusters=5)
clustering.fit(X)
y_clustering = clustering.fit_predict(X)
plt.figure()
for i in range(len(COLORS)):
    plt.scatter(X[y_clustering == i, 0], X[y_clustering == i, 1], color=COLORS[i])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Hierarchial Clustering')
plt.show()
```

### complete method

```python
clustering = AgglomerativeClustering(linkage='complete', n_clusters=5)
clustering.fit(X)
y_clustering = clustering.fit_predict(X)
plt.figure()
for i in range(len(COLORS)):
    plt.scatter(X[y_clustering == i, 0], X[y_clustering == i, 1], color=COLORS[i])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Hierarchial Clustering')
plt.show()
```

```python
clustering = AgglomerativeClustering(linkage='average', n_clusters=7)
clustering.fit(X)
y_clustering = clustering.fit_predict(X)
plt.figure()
for i in range(len(COLORS)):
    plt.scatter(X[y_clustering == i, 0], X[y_clustering == i, 1], color=COLORS[i])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Hierarchial Clustering')
plt.show()
```
